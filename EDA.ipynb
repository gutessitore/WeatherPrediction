```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import warnings
warnings.simplefilter("ignore")
```

```python
df = pd.read_csv('C:/Users/cedua/CDIA - PUCSP/PROJETO 5ยบ SEMESTRE (SAVINO)/WeatherPrediction/data/data.csv')
df.head()
```

```python
#checking null values
for column in df.columns:
    print(column, '---', df[column].isnull().sum(), 'null values')
```

```python
#HISTOGRAMS

columns = [ column for column in df.columns]
del columns[0] #deleting Date --> x axis for all
          
for column in columns:  
    if column == 'Sea-Level Air Pressure':
        fig = px.line(df, x='Date', y=[column],
                 title = column + ' 2021 - 03/2023', 
                 color_discrete_sequence = ['darkcyan'])

        fig.update_layout(yaxis_title = column)

        fig.update_xaxes(rangeslider_visible=True)

        fig.update_yaxes(range=[1000, 1040])

        fig.show()

    else:
        fig = px.line(df, x='Date', y=[column],
                            title = column + ' 2021 - 03/2023', 
                            color_discrete_sequence = ['darkcyan'])

        fig.update_layout(yaxis_title = column)

        fig.update_xaxes(rangeslider_visible=True)

        fig.show()
```

```python
#correlation
sns.heatmap(df.corr(), annot=True);

print("The strongest correlations with Average Temperature ---> Minimun and Maximum Temperature")
```

```python
fig = px.line(df, x='Date', y=['Average Temperature', 'Minimum Temperature', 'Maximum Temperature'],
              color_discrete_sequence=["orange", "blue", "red"], 
              title='Temperature Variation')

fig.update_layout(xaxis_title = 'Date',
                  yaxis_title = 'Temperature (ยบC)' )

fig.update_yaxes(range=[0, 40])
fig.update_xaxes(rangeslider_visible=True)

fig.show()

print('As Average Temperature increases/decreases, Maximum and Minimum Temperature also increases/decreases.')
```
